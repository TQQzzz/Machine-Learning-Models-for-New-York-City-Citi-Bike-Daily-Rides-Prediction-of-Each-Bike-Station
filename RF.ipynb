```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn import svm
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Ridership.csv')
df.head()
```

```python
df.shape
```

```python
df['Unemployment_Rate'] = df['Unemployment_Rate'].str.replace('%','')
df['18-64_Poverty'] = df['18-64_Poverty'].str.replace('%','')
df['Per_Capita_Income'] = df['Per_Capita_Income'].str.replace(',','')
df['Unemployment_Rate'] = pd.to_numeric(df['Unemployment_Rate'], errors='coerce')
df['18-64_Poverty'] = pd.to_numeric(df['18-64_Poverty'], errors='coerce')
df['Per_Capita_Income'] = pd.to_numeric(df['Per_Capita_Income'], errors='coerce')
```

```python
df=df.dropna()
```

```python
df.columns
```

```python
col = df.columns.tolist()
col = [col[1]] + [col[5]] + [col[4]] + col[2:4] + col[6:]
df=df[col]
df.head(3)
```

```python
X = df.iloc[:,3:].values
y = df.iloc[:,2].values

sc_X = StandardScaler()
X = sc_X.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Predict on testing set and print R2
y_pred = rf.predict(X_test)
r2 = r2_score(y_test, y_pred)
print('R2:', r2)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

# Define the parameter grid to search over
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 15],
    'max_features': ['sqrt', 'log2']
}

# Create a random forest regressor
rf = RandomForestRegressor()

# Create a grid search object
grid_search = GridSearchCV(rf, param_grid=param_grid, cv=5)

# Fit the grid search object to the data
grid_search.fit(X_train, y_train)

# Print the best parameters and score
print("Best parameters:", grid_search.best_params_)
print("Best score:", grid_search.best_score_)
```

```python
best_rf = grid_search.best_estimator_

# Use the best model to predict on the test set and calculate the R-squared score
r2 = best_rf.score(X_test, y_test)

# Print the R-squared score as a measure of out-of-sample accuracy
print("R-squared score on test set: {:.4f}".format(r2))
```

# RF with PCA

```python
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
from sklearn.decomposition import PCA

# Apply PCA to the data
pca = PCA()
X_pca = pca.fit_transform(X)

# Calculate the cumulative explained variance ratio
cumulative_var_ratio = np.cumsum(pca.explained_variance_ratio_)

# Plot the cumulative explained variance ratio
plt.plot(cumulative_var_ratio)
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.title('PCA Scree Plot')
plt.show()
```

```python
pca = PCA(n_components=3)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

# Train random forest model
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train_pca, y_train)

# Predict on testing set and print R2
y_pred = rf.predict(X_test_pca)
r2 = r2_score(y_test, y_pred)
print('R2:', r2)
```

```python
import matplotlib.pyplot as plt
# Predict ridership for training and testing sets
y_train_pred = rf.predict(X_train_pca)
y_test_pred = rf.predict(X_test_pca)

# Plot predicted vs actual ridership values for training and testing sets
plt.plot(y_train, label='Actual (Training)')
plt.plot(y_train_pred, label='Predicted (Training)')
plt.plot(y_test, label='Actual (Testing)')
plt.plot(y_test_pred, label='Predicted (Testing)')
plt.xlabel('Observation')
plt.ylabel('Ridership')
plt.title('Random Forest Regression')
plt.legend()
plt.show()
```
